```python
df.drop_duplicates(inplace=True)
```

```python
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].str.replace('Cr', '', regex=False).str.strip().str.replace('Rs','').str.replace('â‚¹','')
df['Ticket_Price_Cr'] = pd.to_numeric(df['Ticket_Price_Cr'], errors='coerce')
```

```python
df['Ticket_Price_Cr']
```

```python
df['Unit_Size_Sqft'] = df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].mean())
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].mean())
df['Amenity_Score'] = df['Amenity_Score'].fillna(df['Amenity_Score'].median())
df['Buyer_Comments'] = df['Buyer_Comments'].fillna("No Comments")
```

```python
df
```

```python
df['Unit_Size_Sqft'] = df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].mean())
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].mean())
df['Amenity_Score'] = df['Amenity_Score'].fillna(df['Amenity_Score'].median())
df['Buyer_Comments'] = df['Buyer_Comments'].fillna("No Comments")
df
```

```python
df.isnull().sum()
```

```python
df['Buyer_Type'] = df['Buyer_Type'].replace({'Other':'OTHERS','Startup Founder':'STARTUP FOUNDER'})
df['Micro_Market'] = df['Micro_Market'].str.title()
df['Possession_Status'] = df['Possession_Status'].str.title()
df['NRI_Buyer'] = df['NRI_Buyer'].str.title()
df['Configuration'] = df['Configuration'].str.upper()
df.rename(columns={"Developer_Name": "Builder"}, inplace=True)
df
```

```python
# Extract quarter column from 'Purchase_Quarter'
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'])
df['Quarter_Number'] = df['Purchase_Quarter'].dt.quarter
df['Quarter_Number']
```

```python
# Calculate Price per Sqft
df["Price_per_Sqft"] = df["Ticket_Price_Cr"] * 10000000 / df["Unit_Size_Sqft"]
df
```

```python
df['Transaction_Type'].unique()
```

```python
# Create Booking_Flag from Transaction_Type
# Let's assume Primary = booked (1), Secondary = not booked (0)
df['Booking_Flag'] = df['Transaction_Type'].str.lower().map({'primary': 1, 'secondary': 0})
```
